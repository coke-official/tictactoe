```
import numpy as np

class TicTacToe:
  
  def __init__(self):
    self.board = np.zeros((3,3), dtype=int)
    self.moves = 0
  
  def move(self):  
    self.moves += 1;
    turn = {1:"o", 0:"x"}
    print(turn[self.moves%2] + "の番です")
    raw = input("行を入力:")
    column = input("列を入力:")
    self.board[int(raw)-1, int(column)-1] = 2-(self.moves%2)

  def show_board(self):
    piece = {0:" ", 1:"o", 2:"x"}
    print(" |1|2|3")
    print("--------")
    for i in range(3):
      raw = str(i+1) + "|" + piece[self.board[i, 0]] + "|" + piece[self.board[i, 1]] + "|" + piece[self.board[i, 2]] + "|"
      print(raw + "\n--------")     
```

```
tic = TicTacToe()
tic.show_board()
for i in range(3):
  tic.move()
  tic.show_board()
```
